These figures, however, do not provide a reliable comparison between companies. To better compare all companies I will plot the Volume * Price.

```python
# Plot 1
Anthem['MarktCap'] = Anthem['Open'] * Anthem['Volume']
Comcast['MarktCap'] = Comcast['Open'] * Comcast['Volume']
BankOfAmerica['MarktCap'] = BankOfAmerica['Open'] * BankOfAmerica['Volume']
Target['MarktCap'] = Target['Open'] * Target['Volume']
Anthem['MarktCap'].plot(label = 'ANTM', figsize = (15,7))
Comcast['MarktCap'].plot(label = "CMCSA")
BankOfAmerica['MarktCap'].plot(label = 'BAC')
Target['MarktCap'].plot(label = "TGT")

plt.title('V * P')
plt.legend()
```

```python
# Plot 2
HP['MarktCap'] = HP['Open'] * HP['Volume']
IBM['MarktCap'] = IBM['Open'] * IBM['Volume']
Cisco['MarktCap'] = Cisco['Open'] * Cisco['Volume']
Merck['MarktCap'] = Merck['Open'] * Merck['Volume']
HP['MarktCap'].plot(label = 'HPE', figsize = (15,7))
IBM['MarktCap'].plot(label = "IBM")
Cisco['MarktCap'].plot(label = 'CSCO')
Merck['MarktCap'].plot(label = "MRK")

plt.title('V * P')
plt.legend()
```

```python
# Plot 3
AbbVie['MarktCap'] = AbbVie['Open'] * AbbVie['Volume']
Progressive['MarktCap'] = Progressive['Open'] * Progressive['Volume']
Nationwide['MarktCap'] = Nationwide['Open'] * Nationwide['Volume']
AmericanExpress['MarktCap'] = AmericanExpress['Open'] * AmericanExpress['Volume']
AbbVie['MarktCap'].plot(label = 'ABBV', figsize = (15,7))
Progressive['MarktCap'].plot(label = "PGR")
Nationwide['MarktCap'].plot(label = 'NFS')
AmericanExpress['MarktCap'].plot(label = 'AXP')

plt.title('V * P')
plt.legend()
```

```python
# Plot 4
USAA['MarktCap'] = USAA['Open'] * USAA['Volume']
CapitalOne['MarktCap'] = CapitalOne['Open'] * CapitalOne['Volume']
HASBRO['MarktCap'] = HASBRO['Open'] * HASBRO['Volume']
Microsoft['MarktCap'] = Microsoft['Open'] * Microsoft['Volume']
Intel['MarktCap'] = Intel['Open'] * Intel['Volume']
USAA['MarktCap'].plot(label = "USISX", figsize = (15,7))
CapitalOne['MarktCap'].plot(label = 'COF')
HASBRO['MarktCap'].plot(label = "HAS")
Microsoft['MarktCap'].plot(label = 'MSFT')
Intel['MarktCap'].plot(label = 'INTC')

plt.title('V * P')
plt.legend()
```
